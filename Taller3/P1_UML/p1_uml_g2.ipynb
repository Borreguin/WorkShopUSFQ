```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import numpy as np

# Función para preparar los datos
def prepare_data():
    _df = pd.read_csv("data/data.csv", sep=';', decimal='.')
    _df['timestamp'] = pd.to_datetime(_df['timestamp'])  # Convertir a formato datetime
    return _df

# Función para agrupar por día y hora
def group_by_day_hour(_df: pd.DataFrame, variable):
    _df['day'] = _df['timestamp'].dt.date  # Extraer la fecha (día)
    _df['hour'] = _df['timestamp'].dt.hour  # Extraer la hora
    grouped = _df.groupby(['day', 'hour'])[variable].mean().unstack()  # Agrupar y calcular el promedio
    
    # Imputar valores NaN con el promedio de la columna
    grouped = grouped.fillna(grouped.mean())
    
    return grouped

# Función para aplicar clustering jerárquico
def apply_agglomerative_clustering(_df: pd.DataFrame, n_clusters=10):
    _df.columns = _df.columns.astype(str)  # Asegurarse de que todos los nombres de las columnas sean strings

    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(_df)  # Normalizar los datos

    agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
    clusters = agg_clustering.fit_predict(scaled_data)
    return clusters

# Función para aplicar k-means
def apply_kmeans(_df: pd.DataFrame, n_clusters=10):
    _df.columns = _df.columns.astype(str)  # Asegurarse de que todos los nombres de las columnas sean strings

    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(_df)  # Normalizar los datos

    kmeans = KMeans(n_clusters=n_clusters, random_state=2204)
    clusters = kmeans.fit_predict(scaled_data)
    return clusters, kmeans.cluster_centers_  # Retornar los centroides

# Función para graficar los datos por días y horas con clustering
def plot_days_hours_line_with_clusters(_df: pd.DataFrame, variable, clusters, title, n_clusters):
    plt.figure(figsize=(10, 4))
    
    # Crear un colormap dinámico basado en el número de clusters
    #cmap = plt.cm.get_cmap('Set1', n_clusters)  # 'Set1' es una buena paleta para múltiples categorías
    cmap = plt.get_cmap('Set1', n_clusters)  # 'Set1' es una buena paleta para múltiples categorías

    for i, day in enumerate(_df.index):
        # Usamos el colormap para asignar colores según el cluster
        plt.plot(_df.columns, _df.loc[day], label=day, color=cmap(clusters[i] / (n_clusters - 1)))

    plt.title(f'{title} de {variable} por Hora y Día')
    plt.xlabel('Hora del día')
    plt.ylabel(variable)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Función para contar cuántas líneas pertenecen a cada cluster y mostrar los días (máximo 10 días por cluster)
def show_days_per_cluster(clusters, grouped_df, method_name, max_days=10):
    # Crear un DataFrame con los días correspondientes a cada cluster
    cluster_days = pd.DataFrame({'cluster': clusters})
    cluster_days['day'] = grouped_df.index  # Asegúrate de que el índice contenga los días

    clusters_grouped = cluster_days.groupby('cluster')['day'].apply(lambda x: x.unique().tolist())  # Agrupar días por cluster

    print(f"Días correspondientes a cada cluster ({method_name}):")
    for cluster_id, days in clusters_grouped.items():
        if len(days) > max_days:
            print(f"Cluster {cluster_id}: {', '.join([str(day) for day in days[:max_days]])}... y {len(days) - max_days} días más.")
        else:
            print(f"Cluster {cluster_id}: {', '.join([str(day) for day in days])}")
    print("\n")

# Función para graficar los centroides de los clusters
def plot_centroids(centroids, title):
    plt.figure(figsize=(10, 4))
    
    # Graficar cada centroide
    for i, centroid in enumerate(centroids):
        plt.plot(np.arange(len(centroid)), centroid, label=f'Cluster {i}')
    
    plt.title(f'Centroides de {title}')
    plt.xlabel('Hora del día')
    plt.ylabel('Valor promedio')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Función para calcular centroides para clustering jerárquico
def calculate_hierarchical_centroids(_df: pd.DataFrame, clusters, n_clusters):
    centroids = []
    for cluster in range(n_clusters):
        cluster_data = _df[clusters == cluster]
        centroid = cluster_data.mean(axis=0)
        centroids.append(centroid)
    return np.array(centroids)

# Código principal
if __name__ == "__main__":
    # Preparar datos
    df = prepare_data()

    # Agrupar los datos por día y hora
    grouped = group_by_day_hour(df, 'V005_vent01_CO2')
    
    # Aplicar k-means
    clusters_kmeans, centroids_kmeans = apply_kmeans(grouped, n_clusters=15)
    plot_days_hours_line_with_clusters(grouped, 'V005_vent01_CO2', clusters_kmeans, 'Clusters K-means', n_clusters=3)
    show_days_per_cluster(clusters_kmeans, grouped, 'K-means')
    
    # Graficar los centroides de K-means
    plot_centroids(centroids_kmeans, 'K-means')

    # Aplicar clustering jerárquico
    clusters_agglomerative = apply_agglomerative_clustering(grouped, n_clusters=15)
    plot_days_hours_line_with_clusters(grouped, 'V005_vent01_CO2', clusters_agglomerative, 'Clusters Jerárquicos', n_clusters=3)
    show_days_per_cluster(clusters_agglomerative, grouped, 'Jerárquico')

    # Calcular y graficar los centroides del clustering jerárquico
    centroids_agglomerative = calculate_hierarchical_centroids(grouped, clusters_agglomerative, n_clusters=15)
    plot_centroids(centroids_agglomerative, 'Jerárquico')
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.impute import SimpleImputer
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np

# Asumiendo que ya tienes el DataFrame df y la columna 'timestamp'
features = ['V005_vent01_CO2', 'V006_vent01_temp_out']
X = df[features]

# Imputar valores faltantes usando la media de cada columna
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Normalizar los datos para que los valores tengan una escala comparable
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

### PARTE 1: K-MEANS ###
# Aplicar K-Means con un número arbitrario de clústeres, por ejemplo, 40
kmeans = KMeans(n_clusters=12, random_state=2204)
df['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

# Extraer los centroides del K-Means
centroides_kmeans = kmeans.cluster_centers_

# Visualizar los centroides de K-Means
plt.figure(figsize=(12, 8))
plt.scatter(centroides_kmeans[:, 0], centroides_kmeans[:, 1], c='red', s=200, marker='X', label='Centroides K-Means')  # Usa un marcador distintivo como una 'X'
plt.xlabel('V005_vent01_CO2')
plt.ylabel('V006_vent01_temp_out')
plt.title('Centroides de Clustering usando K-Means')
plt.grid(True)
plt.legend()
plt.show()

### PARTE 2: GMM ###
# Aplicar GMM para el clustering
gmm = GaussianMixture(n_components=11, random_state=2204)
df['Cluster_GMM'] = gmm.fit_predict(X_scaled)

# Obtener los centroides (medias de las componentes gaussianas)
centroides_gmm = gmm.means_

# Visualizar los centroides de GMM
plt.figure(figsize=(12, 8))
plt.scatter(centroides_gmm[:, 0], centroides_gmm[:, 1], c='blue', s=200, marker='o', label='Centroides GMM')
plt.xlabel('V005_vent01_CO2')
plt.ylabel('V006_vent01_temp_out')
plt.title('Centroides de Clustering usando GMM')
plt.grid(True)
plt.legend()
plt.show()

### PARTE 3: Método del Codo para K-Means ###
def plot_elbow_method(X):
    distortions = []
    K = range(1, 20)  # Probamos desde 1 hasta 20 clusters
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=2204)
        kmeans.fit(X)
        distortions.append(kmeans.inertia_)  # Inercia dentro de los clusters

    # Gráfico de codo
    plt.figure(figsize=(8, 6))
    plt.plot(K, distortions, 'bx-')
    plt.xlabel('Número de clusters')
    plt.ylabel('Distorsión')
    plt.title('Método del codo para determinar el número óptimo de clusters')
    plt.grid(True)
    plt.show()

# Mostrar el gráfico de codo con los datos escalados
plot_elbow_method(X_scaled)

### PARTE 4: BIC para GMM ###
def plot_bic_gmm(X):
    bics = []
    K = range(1, 20)
    for k in K:
        gmm = GaussianMixture(n_components=k, random_state=2204)
        gmm.fit(X)
        bics.append(gmm.bic(X))  # Cálculo del BIC

    # Gráfico de BIC
    plt.figure(figsize=(8, 6))
    plt.plot(K, bics, 'bo-')
    plt.xlabel('Número de componentes')
    plt.ylabel('BIC')
    plt.title('Criterio BIC para determinar el número óptimo de componentes en GMM')
    plt.grid(True)
    plt.show()

# Mostrar el gráfico de BIC con los datos escalados
plot_bic_gmm(X_scaled)

### PARTE 5: Dendrograma de los centroides de K-Means ###
# Utilizar clustering jerárquico en los centroides de K-Means
Z = linkage(centroides_kmeans, method='ward')  # Usamos el método 'ward' para minimización de varianza

# Graficar el dendrograma
plt.figure(figsize=(10, 8))
dendrogram(Z)
plt.title('Dendrograma de los centroides de K-Means')
plt.xlabel('Índice del clúster')
plt.ylabel('Distancia')
plt.show()

Z = linkage(centroides_gmm, method='ward')  # Usamos el método 'ward' para minimización de varianza

# Graficar el dendrograma
plt.figure(figsize=(10, 8))
dendrogram(Z)
plt.title('Dendrograma de los centroides de K-Means')
plt.xlabel('Índice del clúster')
plt.ylabel('Distancia')
plt.show()
```
